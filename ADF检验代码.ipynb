```python
# import xlrd
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
import pandas as pd
import numpy as np
import datetime 

endtime = datetime.datetime.now()
starttime = datetime.datetime.now()-datetime.timedelta(minutes=5)
table = pd.read_excel(r"D:\software_bag_X\MicrosoftVScodefiles\ADF预测数据.xlsx",engine='openpyxl')
for i in range(2,len(table)):
    x_data=[]
    y_data=[]
    cap = table.iloc[:,i]
    cap1 = table.iloc[:,0]
    print(cap,cap1)
    for i in range(1,len(cap)-1):
        x_data.append(cap1[i])
        y_data.append(cap[i])

    adf_seq = np.array(y_data)
    dftest = adfuller(adf_seq,autolag='AIC')
    if dftest[1]>0.05:
        adf_seq1 = np.array(pd.DataFrame(y_data).diff(periods=1))
        print(adf_seq1[1:])
        dftest = adfuller(adf_seq1[1:],autolag='AIC')
        if dftest[1]>0.05:
            print(starttime)
            print(endtime)
            print(dftest)
            print(f"当前{cap1.name}还是不平稳")
        else:
            print(dftest[1])
            print(starttime)
            print(endtime)
            print(dftest)
            print(f"当前{cap1.name}平稳")
    else:
        print(dftest[1])
        print(starttime)
        print(endtime)
        print(dftest)
        print(f"当前{cap1.name}平稳")

    #画折线图
    #用来正常显示中文标签
    plt.rcParams['font.sans-serif']=['SimHei'] 
    plt.plot(x_data, y_data, 'g-s')
    plt.title('data')
    plt.legend()

plt.xlabel('id')
plt.ylabel('data')
plt.show()
```
